```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_line_scores(scores, title="Desempenho dos Scores", xlabel="Execução", ylabel="Score"):
    """
    Plota um gráfico de linhas para um conjunto de scores, incluindo linhas de máximo, média e mínimo.
    
    Parâmetros:
    - scores: lista ou array com os valores dos scores.
    - title: título do gráfico.
    - xlabel, ylabel: rótulos dos eixos.
    """
    plt.figure(figsize=(10,6))
    plt.plot(range(1, len(scores)+1), scores, marker='o', linestyle='-', color='blue', label='Score')
    
    max_score = max(scores)
    min_score = min(scores)
    mean_score = sum(scores) / len(scores)
    
    plt.axhline(max_score, color='red', linestyle='--', label=f'Máximo ({max_score:.2f})')
    plt.axhline(mean_score, color='orange', linestyle='-.', label=f'Média ({mean_score:.2f})')
    plt.axhline(min_score, color='green', linestyle=':', label=f'Mínimo ({min_score:.2f})')
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()
```

```python
def plot_bar_distribution(scores, bins=10, title="Distribuição dos Scores", xlabel="Score", ylabel="Frequência"):
    """
    Plota um histograma em formato de barras para mostrar a distribuição dos scores e inclui a linha da média.
    
    Parâmetros:
    - scores: lista ou array com os valores dos scores.
    - bins: número de intervalos do histograma.
    - title, xlabel, ylabel: títulos e rótulos.
    """
    plt.figure(figsize=(10,6))
    sns.histplot(scores, bins=bins, kde=False, color='green', edgecolor='black')
    mean_score = sum(scores) / len(scores)
    plt.axvline(mean_score, color='orange', linestyle='--', linewidth=2, label=f'Média ({mean_score:.2f})')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()
```

```python
scores = [181.38, 110.28, 27.96, 29.48, 20.21, 19.32, 109.29, 79.83, 106.41, 66.64, 41.99, 42.35, 142.68, 21.99, 105.24, 106.50, 117.57, 44.35, 8.82, 115.68, 118.29, 44.68, 109.20, 69.60, 119.28, 108.12, 44.50, 105.15, 110.55, 111.36]
plot_line_scores(scores, title="Linha dos Scores do Agente")
plot_bar_distribution(scores, bins=5, title="Distribuição dos Scores do Agente")
```
